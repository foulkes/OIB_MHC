#### <span style="color:darkblue"> B.1 Numerical data 


#### <span style="color:darkblue"> Summary measures:


The collection of values for any one of the numerical, continuous variables (`weight`, say) is the <span style="color:orange">distribution <span style="color:black">for that variable.  

Numerical and graphical summaries provide information about a distribution without listing all the values

Numerical measures

- Measures of center:  <span style="color:orange">mean, median 

- Measures of spread: <span style="color:orange"> variance, standard deviation, first and third quartiles, interquartile range 


#### Measures of center:####

- The <span style="color:orange">sample mean <span style="color:black"> of a variable is the sum of all observations divided by the number of observations:

$$\overline{x} = \frac{x_1+x_2+\cdots+x_n}{n}$$  
where $x_1, x_2, \ldots, x_n$ represent the $n$ observed values.

Simple command in R to calculate mean:

```r
require(dplyr)
summarize(famuss, mean(weight))
```

- The <span style="color:orange">median <span style="color:black"> is the value of the middle observation in a  dataset.  

If the number of observations is 

- Odd, the median is the middle observation
- Even, the median is the average of the two middle observations

Approximately 50% of the observations lie below (or above) the median


```r
summarize(famuss,median(weight))
```

In *FAMuSS*,  mean and median weight are not equal.  Explanation coming later...

#### Measures of spread

The distance $x - \overline{x}$ of an observation $x$ from the mean $\overline{x}$ is its <span style="color:black">deviation <span style="color:black">.
- The <span style="color:orange">sample variance <span style="color:black"> the sum of squared deviations divided by the number of observations minus 1.
$$s^2 = \frac{({x_1 - \overline{x})}^{2}+({x_2 - \overline{x})}^{2}+\cdots+({x_n - \overline{x})}^{2}}{n-1}$$

where $x_1, x_2, \dots, x_n$ represent the $n$ observed values.

- The <span style="color:orange">standard deviation <span style="color:black"> ($s$) is the square root of the variance
$$s = \sqrt{\frac{({x_1 - \overline{x})}^{2}+({x_2 - \overline{x})}^{2}+\cdots+({x_n - \overline{x})}^{2}}{n-1}}$$

- The <span style="color:orange">first quartile<span style="color:black"> ($Q_1$) is the $25^{th}$ percentile of a dataset: 25% of the observations fall below this value.

  - Calculate it as the median of the values smaller than the median.


- The <span style="color:orange">third quartile<span style="color:black"> ($Q_3$) is the $75^{th}$ percentile of a dataset: 75% of the observations fall below this value.

  - Calculate it as the median of the values larger than the median.


- The <span style="color:orange">interquartile range <span style="color:black">(IQR) is the distance between the third and first quartile: $\text{IQR} = Q_3 - Q_1$

*(Note: In small datasets, software may produce slightly different values; ignore those differences.)*


In R we calculate these summary measures as follows:

```r
summarize(famuss,sd(weight))
```

```r
famuss %>% 
    summarize(sd(weight))
```

```r
famuss %>% 
    summarize(min=min(weight), max=max(weight), q1=quantile(weight,p=.25), q3=quantile(weight,p=.75))
```

The median and IQR are called <span style="color:orange">robust statistics <span style="color:black"> because they are less likely to be affected by extreme values.

Now let's consider how to visualize the relationship between two numerical variables...

#### Scatterplot

```r
ggplot(data=famuss,aes(x=height, y=weight))+
    geom_point(alpha = 0.4)
```

We see from this figure that height and weight are <span style="color:orange">correlated


#### Correlation

Two variables $x$ and $y$ are 

- *Positively correlated* if $y$ increases as $x$ increases 

- *Negatively correlated* if $y$ decreases as $x$ increases 

Height and weight are _**positively**_ correlated

Correlation is discussed in Section 1.4.5 in *OI Biostat*

Details not needed for now; will be covered when we discuss regression in Unit 6.  


#### Correlation can be measured with the correlation coefficient (r)

The correlation between two variables $x$ and $y$ is given by:
		$$r =  \frac{1}{n-1}\sum^{n}_{i=1}
          \left(\frac{x_{i}-\overline{x}}
          {s_{x}}\right)\left(\frac{y_{i}-\overline{y}}{s_{y}}\right)$$
          
$(x_1,y_1), (x_2,y_2), \ldots, (x_n, y_n)$ are the $n$ paired values of $x$ and $y$.

$s_x=\text{sd}(x), \,\, s_y = \text{sd}(y)$ 
        
A consequence of the formula: $-1 \leq r \leq 1$


We can visualize this correlation by adding a "best fit" straight line...

```r
ggplot(data=famuss,aes(x=height, y=weight))+
    geom_point(alpha = 0.4) +
    geom_smooth(method="lm",colour="darkblue",se=FALSE)
```

#### <span style="color:darkblue"> B.2 Categorical data 

#### Summary measures 

Numerical summaries cannot be used with categorical data  

- <span style="color:orange">Contingency tables <span style="color:black"> summarize the distribution of categories
- Also called <span style="color:orange">*frequency tables*

What can be learned about the genotype of the study participants at the location *actn3.r577x*?

#### A simple contingency table


```r
famuss %>%
    count(actn3.r577x)
```

This table is called a one-way table.  


We can also looks at the counts across levels of two categorical variables. To see this, we will first create a new variable called `bmi25` that equals TRUE if body mass index (BMI) is greater than 25 and FALSE if BMI is less than 25. We do this with the `mutate()` function:

```r
famuss = famuss %>%
    mutate(bmi25 = bmi>25)
head(famuss)
```

Using the `count()` function, we create a new variable with the name `n` that gives the counts across levels of the indicated categorical variables:

```r
famuss %>%
    count(bmi25,actn3.r577x)
```

There are lots of different ways to do the same thing in R!  For example, we could have instead used the `group_by()` and `summarize()` functions to get the same result:

```r
famuss %>% 
    group_by(bmi25,actn3.r577x) %>%
    summarize(n=n())
```

This can be made into a <span style="color:orange">two-way contingency table <span style="color:black"> using the `spread()` function:

```r
require(tidyr)
famuss %>%
    count(bmi25,actn3.r577x) %>%
    spread(key=actn3.r577x,value=n)
```

#### <span style="color:darkblue"> Exploring categorical data graphically:
- Barplots (segmented)

A barplot for a single categorical variable:

```r
ggplot(data=famuss,aes(actn3.r577x)) +
    geom_bar(alpha=.4)
```

Similar to what we saw with numerical data, we can visualize data on a categoical variables over levels of another categorical variable.  For example, here is a <span style="color:orange">segmented bar plot<span style="color:black">:

```r
ggplot(data=famuss,aes(actn3.r577x,fill=bmi25)) +
    geom_bar(alpha=.6)
```

There are ***at least three other ways*** we can visualize these data using bar plots:

```r
ggplot(data=famuss,aes(actn3.r577x,fill=bmi25)) +
    geom_bar(alpha=.6,position="dodge")
```

```r
ggplot(data=famuss,aes(actn3.r577x,fill=bmi25)) +
    geom_bar(alpha=.6,position="fill")
```

```r
ggplot(data=famuss,aes(bmi25,fill=actn3.r577x)) +
    geom_bar(alpha=.6,position="fill")
```
